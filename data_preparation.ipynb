```python
# import module 
from geopy.geocoders import Nominatim 

# initialize Nominatim API 
geolocator = Nominatim(user_agent="geoapiExercises") 
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
columnsl = []
for i in range(1,69):
    columnsl.append('f' + str(i))
columnsl.append('latitude')
columnsl.append('longitude')
columnsa = np.array(columnsl)
columnsl2 = []
for i in range(1,117):
    columnsl2.append('mcff' + str(i))
columnsl2.append('latitude')
columnsl2.append('longitude')
columnsa2 = np.array(columnsl2)
```

```python
df = pd.read_csv('Geographical Original of Music/default_features_1059_tracks.csv',header=None)
df.dropna(inplace=True)
df.columns=columnsa
```

```python


def find_country(Latitude, Longitude):
    location = geolocator.reverse(Latitude+","+Longitude, language='en') 
    address = location.raw['address'] 
    country = address.get('country', '')
    print(country)
    return country
```

```python
lat = str(df.iloc[586]['latitude'])
long = str(df.iloc[586]['longitude'])
find_country(lat,long)
```
